# Visualization

1. Wind Speed Distribution
    - Create a histogram or boxplot of Wind Speed_km/h to understand its distribution and identify outliers.

```python
plt.figure(figsize=(10,5))
sns.histplot(data['Wind Speed_km/h'], bins=30, kde=True)
plt.title('Distribution of Wind Speed')
plt.xlabel('Wind Speed (km/h)')
plt.savefig("Figures/wind_speed_distribution.png")
plt.show()
```

2. Relationship Between Temperature and Dew Point
    - Use a scatter plot to explore how temperature relates to dew point temperature.

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='Temp_C', y='Dew Point Temp_C', data=data)
plt.title('Temperature vs Dew Point Temperature')
plt.xlabel('Temperature (°C)')
plt.ylabel('Dew Point Temperature (°C)')
plt.savefig("Figures/temp_vs_dew_point.png")
plt.show()
```

3. Average Visibility for Each Weather Condition
    - Use a bar plot to show mean Visibility_km for each unique weather condition.

```python
plt.figure(figsize=(12,15))
mean_visibility = data.groupby('Weather Condition')['Visibility_km'].mean().sort_values()
mean_visibility.plot(kind='barh', color='skyblue')
plt.title('Average Visibility by Weather Condition')
plt.xlabel('Visibility (km)')
plt.ylabel('Weather Condition')
plt.savefig("Figures/avg_visibility_by_weather_condition.png")
plt.show()
```

4. Boxplots of Temperature Across Different Weather Conditions
    - Compare the temperature distribution across different weather types.

```python
plt.figure(figsize=(20,12))
sns.boxplot(x='Weather Condition', y='Temp_C', data=data)
plt.title('Temperature Distribution by Weather Condition')
plt.xticks(rotation=90)
plt.xlabel('Weather Condition')
plt.ylabel('Temperature (°C)')
plt.savefig("Figures/temp_distribution_by_weather_condition.png")
plt.show()
```

5. Wind Speed vs Visibility for High Wind Speeds
    - Create a scatter plot for instances where wind speed is above a certain threshold (e.g., > 24 km/h), to see if visibility is affected.

```python
high_wind = data[data['Wind Speed_km/h'] > 24]

plt.figure(figsize=(8,6))
sns.scatterplot(x='Wind Speed_km/h', y='Visibility_km', data=high_wind)
plt.title('Visibility vs High Wind Speeds (>24 km/h)')
plt.xlabel('Wind Speed (km/h)')
plt.ylabel('Visibility (km)')
plt.savefig("Figures/wind_speed_vs_visibility_high.png")
plt.show()
```
